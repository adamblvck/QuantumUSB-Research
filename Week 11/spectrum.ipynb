```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

%matplotlib inline

SAMPLE_NUMBER = 213

import os
import time
```

```python
def read_filename(filename):
	
	# read table
	df = pd.read_csv(filename, sep='\t', decimal='.', header=0)

	# rename columns, and make sure we only keep the two that we're interested in
	df.columns = ['Wavelength', 'PL (Counts/s)']
	df = df[['Wavelength', 'PL (Counts/s)']]

	# convert column values to float
	df['Wavelength'] = df['Wavelength'].astype(float)
	df['PL (Counts/s)'] = df['PL (Counts/s)'].astype(float)

	return df
```

```python
df_spectrum = read_filename("./Spectrum/PL spectrum center 03-03-2023 545 nm 0.2V.csv")
```

```python
df_spectrum = df_spectrum.loc[df_spectrum['PL (Counts/s)'] < 1000,:]
```

```python
fig, (ax1) = plt.subplots(1,1, figsize=(10,4))  # 1 row, 2 columns

df_spectrum.plot(x='Wavelength', y='PL (Counts/s)', color='#369F0B', ax=ax1, title=f"#{SAMPLE_NUMBER} ODMR - 0.3V Microwave - 0V Potential")
ax1.grid(True)
```
